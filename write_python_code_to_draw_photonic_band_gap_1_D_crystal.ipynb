Sure, here is a Python code to draw the photonic band gap of a 1D crystal:

```
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the system
a = 500  # lattice constant (nm)
n = 3.4  # refractive index of the material

# Calculate the Brillouin zone
k = np.linspace(-np.pi / a, np.pi / a, 1000)

# Calculate the photonic band structure
E = np.zeros((1000, 1))
for i in range(len(k)):
    E[i, 0] = n * np.sqrt(k[i]**2)

# Plot the photonic band structure
plt.figure()
plt.plot(k, E)
plt.xlabel("k (1/nm)")
plt.ylabel("Energy (eV)")
plt.title("Photonic band structure of a 1D photonic crystal")
plt.show()
```

This code will plot the photonic band structure of a 1D photonic crystal with a lattice constant of 500 nm and a refractive index of 3.4. The photonic band structure is plotted as a line, with the energy of the modes increasing from bottom to top. The Brillouin zone is plotted as a vertical dashed line.
